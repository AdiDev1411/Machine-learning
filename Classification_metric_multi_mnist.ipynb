```python
import pandas as pd
```

```python
df = pd.read_csv('digit-recognizer/train.csv')
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(x_train,y_train)
```

```python
clf2.fit(x_train,y_train)
```

```python
y_pred1 = clf1.predict(x_test)
```

```python
y_pred2 = clf2.predict(x_test)
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

```python

print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

```python

print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision_score(y_test,y_pred2,average='weighted')
```

```python
recall_score(y_test,y_pred2,average='weighted')
```

```python
f1_score(y_test,y_pred2,average='weighted')
```

```python

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred2))
```
