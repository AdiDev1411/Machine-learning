```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
df
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
df['GarageQual'].mode()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax)
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax)
```

```python
temp = df[df['GarageQual']=='TA']['SalePrice']
```

```python
df['GarageQual'].fillna(df['GarageQual'].mode,inplace=True)
df.isnull().sum()
```

```python


df['GarageQual_mode'] = df['GarageQual']
df['GarageQual_mode'].fillna(df['GarageQual_mode'].mode)

```

```python

df['GarageQual_mode'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde')
df[df['GarageQual_mode']=='TA']['SalePrice'].plot(kind='kde')
```
