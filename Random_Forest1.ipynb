```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_classification
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.tree import plot_tree
```

```python
X , y = make_classification(n_samples=500, n_features=2 , n_informative=2 , n_classes=2 ,n_redundant=0 ,random_state=2)
```

```python
df = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(X.shape[1])])
df['Target'] = pd.DataFrame(y)

df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
```

```python
y.shape
```

```python
plt.scatter(X[:,0] ,X[:,1] , c =y )
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X , y)
```

```python
plot_tree(dt)
```

```python
plt.figure(figsize=(10, 6))
plot_tree(dt, filled=True, feature_names=[f'Feature_{i}' for i in range(X.shape[1])])
plt.show()
```

```python
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z = dt.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.3)  # Decision regions

# Step 4: Scatter actual points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=40)
plt.title("Decision Tree Classification Boundaries")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rt = RandomForestClassifier(n_estimators=500, random_state=42)
```

```python
rt.fit(X , y)
```

```python
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z = rt.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.3)  # Decision regions

# Step 4: Scatter actual points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=40)
plt.title("Decision Tree Classification Boundaries")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.grid(True)
plt.show()
```
