# Import


```python
import pandas as pd

```

```python
df = pd.read_csv(r'wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class lable','Alcohole','Malic Acid']
df
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=df['Alcohole'],y=df['Malic Acid'],hue=df['Class lable'])
```

# Train and test 

```python
x= df[['Alcohole','Malic Acid']]
y=df['Class lable']
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(x_train)
```

```python
x_train_scalered = scaler.transform(x_train)
x_test_scalered = scaler.transform(x_test)
```

```python
x_train_scalered = pd.DataFrame(x_train_scalered,columns=x_train.columns)

x_test_scalered = pd.DataFrame(x_test_scalered,columns=x_test.columns)

```

# Before Vs After Normalization

```python
import matplotlib.pyplot as plt
```

```python
fig , (ax1,ax2) = plt.subplots(ncols=2,figsize = (12,6))
ax1.set_title("Before Scalling")
ax2.set_title("After Scalling")
sns.scatterplot(x=x_train['Alcohole'],y=x_train['Malic Acid'],c=y_train,ax=ax1)
sns.scatterplot(x=x_train_scalered['Alcohole'],y=x_train_scalered['Malic Acid'],c=y_train,ax=ax2)
plt.show
```

```python
fig , (ax1,ax2) = plt.subplots(ncols=2,figsize = (12,6)) 
ax1.set_title("Before Scalling")
ax2.set_title("After Scalling")
sns.kdeplot(x_train,ax=ax1)
sns.kdeplot(x_train_scalered,ax=ax2)
plt.show()
```
