# **Histrogram of each Colums**

```python
df.shape[1]
```

```python
for column in df.columns:
    plt.figure(figsize=(4, 4))
    plt.hist(df[column].dropna(), bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
