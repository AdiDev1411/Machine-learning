# Import

```python
import pandas as pd
```

```python
df = pd.read_csv(r'Social_Network_Ads.csv')
# For selction of row 
df = df.iloc[:,2:]

df.sample(5)


```

# Train and Tast before standardization

```python
X = df[['Age','EstimatedSalary']]
Y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0) 

```

# Standerdzation

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

```

* Learn the model

```python
scaler.fit(x_train)
```

* Transform the values

```python
x_train_scaler=scaler.transform(x_train)
x_test_scaler=scaler.transform(x_test)
```

```python
scaler.mean_
```

* convert into dataframe

```python
x_train_scaler = pd.DataFrame(x_train_scaler,columns=x_train.columns)
```

```python
x_test_scaler = pd.DataFrame(x_test_scaler,columns=x_test.columns)
```

```python
import numpy as np
```

```python
np.round(x_train_scaler.describe(),5)
```

# Plot a graph

```python
import matplotlib.pyplot as plt
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,6))

ax1.set_title("Before scaling")
ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])


ax2.set_title("After scaling")
ax2.scatter(x_train_scaler['Age'],x_train_scaler['EstimatedSalary'],c='red')
plt.show()
```

* Showing diffrermce

```python
import seaborn as sns

fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,6))

#before scaling
ax1.set_title("Before scaling")
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)


#after scalling 
ax2.set_title("After Scalling")
sns.kdeplot(x_train_scaler['Age'],ax=ax2)
sns.kdeplot(x_train_scaler['EstimatedSalary'],ax=ax2)
```
